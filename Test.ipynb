```python
import xml.etree.ElementTree as ET
import glob
from collections import Counter
import matplotlib.pyplot as plt
import os
```

```python
def get_tag_info(xml_string, tag):
    root = ET.fromstring(xml_string)
    elements = root.findall(f'.//{tag}')
    tag_info = [elem.text for elem in elements if elem.text is not None]

    return tag_info
```

```python
xml_files = glob.glob('Datasets/annotations/*.xml')

print(len(xml_files))
```

```python
data = []
for file in xml_files:
    temp = get_tag_info(open(file).read(), 'name')
    for i in temp: data.append(i)
    
print(data)
```

```python
count_data = Counter(data)
print(count_data)
```

```python
plt.figure(figsize=(8, 6))
plt.bar(count_data.keys(), count_data.values(), color=['blue', 'green', 'red'])
plt.xlabel('Label Type')
plt.ylabel('Count')
plt.title('Distribution of Labels within Dataset1')
plt.show()
```

```python
def get_data_names(files):
    temp = [file.split('.')[0] for file in files]
    return [file.split('\\')[1] for file in temp]
```

```python
print(get_data_names(xml_files))
```
